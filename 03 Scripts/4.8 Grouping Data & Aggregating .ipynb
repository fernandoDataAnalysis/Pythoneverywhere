# loyalty flag for existing customers using the transform() and loc() functions.

```python
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
ords_prods_merge.head(15)
```

```python
ords_prods_merge.sort_values('user_id')
```

# Creating a Loyalty Flag

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
ords_prods_merge[['user_id', 'loyalty_flag', 'order_number']].head(60)
```

# check the basic statistics of the product prices for each loyalty category (Loyal Customer, Regular Customer, and New Customer)

```python
ords_prods_merge.groupby('loyalty_flag')
```

```python
ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

# creating a spending flag based on the average price across all their orders

```python
ords_prods_merge['price_mean'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge.head(5)
```

```python
ords_prods_merge.loc[ords_prods_merge['price_mean'] < 10, 'spending'] = 'Low Spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['price_mean'] >= 10, 'spending'] = 'High Spender'
```

```python
ords_prods_merge.head()
```

```python
ords_prods_merge.sort_values(by='user_id')
```

# order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column

```python
ords_prods_merge['order_frequency_mean'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.mean)
```

```python
ords_prods_merge.loc[ords_prods_merge['order_frequency_mean'] > 20, 'order_frequency'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['order_frequency_mean'] > 10) & (ords_prods_merge['order_frequency_mean']<= 20), 'order_frequency'] = 'Regular Customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['order_frequency_mean'] <= 10, 'order_frequency'] = 'Frequent customer'
```
